```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from cleaning_function import *
from oscars_function import *
```

```python
df_IMDb = cleaning_function()
df_oscars = oscars_function()
```

```python
df_IMDb = cleaning_function()
df_IMDb["Year"] = df_IMDb["Year"].str.replace('TV Movie 2019', '2019')
df_IMDb.drop(df_IMDb[df_IMDb.Year.isnull()].index, inplace = True)
df_IMDb["Year"] = df_IMDb["Year"].astype(int)
df_IMDb = df_IMDb.sort_values("Avg_vote",ascending=False)
df_IMDb
```

```python
df_oscars = df_oscars.sort_values("Year")
df_oscars.Year = df_oscars.Year.str.replace('[/]', '')
df_oscars["Year"] = df_oscars["Year"].astype(int)
df_oscars = df_oscars[df_oscars["Year"] <= 2019]
df_oscars
```

```python
merge = pd.merge(df_IMDb, df_oscars, on=["Film", "Year"])
merge
```
