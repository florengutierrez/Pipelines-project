```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### En este scatterplot podemos comprobar que existe una correlación entre las obras con mas galardones y las valoraciones de IMDb.

```python
df = pd.read_csv("./merge.csv", encoding = "latin-1")
sns.scatterplot(y="Awards", x="Avg_vote", data = df)
plt.savefig('IMDb_&_Oscars.png')
```
